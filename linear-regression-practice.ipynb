# Simple Linear Regression

```python
import pandas as pd
import numpy as np
x = np.linspace(0,1,20) #including 0 and 1
m = 2
c = 0.5
y = m * x + c
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.axis([-0.1,1.1,0,3]) #x_lower,x_upper,y_lower,y_upper
plt.grid()
```

```python
x.reshape(-1,1) #-1 is rows, 1 is column
```

```python
x.reshape(1,-1)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x.reshape(-1,1),y)
```

```python
yhat = lr.predict(x.reshape(-1,1))
yhat
```

```python
plt.scatter(x,y,c='b',label='Ground Truth')
plt.plot(x,yhat,'g-',label='sklearn')
plt.axis([-0.1,1.1,0,3])
plt.legend()
plt.grid()
```

```python
lr.coef_
```

```python
lr.intercept_
```
