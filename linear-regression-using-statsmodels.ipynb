```python
from sklearn import datasets as data
boston = data.load_boston()
df = pd.DataFrame(boston.data,columns=boston.feature_names)
df['value'] = boston.target
df.head()
```

```python
print(boston.DESCR)
```

```python
from sklearn.linear_model import LinearRegression
X = boston.data
y = boston.target
lr = LinearRegression()
lr.fit(X,y)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
yhat = lr.predict(X)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
print(r2_score(y,yhat))
print(mean_squared_error(y,yhat))
```

# Using Statsmodels

```python
import statsmodels.api as sm
model = sm.OLS(y,X)    #(y,X) instead of (X,y)
result = model.fit()
result.summary()
```

```python
#We forgot to add the constant.
#Now we will add constant
import statsmodels.api as sm
X_train = sm.add_constant(X)
model = sm.OLS(y,X_train)
result = model.fit()
result.summary()
```

# Check for VIF for multicollinearity

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif = pd.DataFrame()
collist = boston.feature_names
print(collist)
```

```python
vif['features'] = collist
```

```python
X = df[collist].values
y = df['value'].values
print(X)
```

```python
X.shape[1]
```

```python
vif['VIF'] = [variance_inflation_factor(X,i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'],2)
vif = vif.sort_values(by="VIF",ascending = False)
vif
```

# Now trying Recursive Feature Elimination and building model

```python
from sklearn.feature_selection import RFE
X = boston.data
y = boston.target
lr = LinearRegression()
rfe = RFE(lr,step=1,verbose=1)
x_red = rfe.fit_transform(X,y)

lr.fit(x_red,y)
print(lr.coef_)
print(lr.intercept_)
yhat = lr.predict(x_red)
from sklearn.metrics import r2_score,mean_squared_error
print(r2_score(y,yhat))
print(mean_squared_error(y,yhat))
print(lr.score(x_red,y))
```
